```python
import os
from scipy.fft import fft
from scipy.io import loadmat
import numpy as np
import pandas as pd
```

## Get FFT features for all the patients

```python
target_length = 50


def perform_fft(signal):
    fft_result = fft(signal, axis=-1)
    amplitude = np.abs(fft_result)
    result_length = amplitude.shape[1]
    step_size = int((result_length // 2) // target_length)
    return amplitude[:, :(result_length // 2)][:, ::step_size]
```

```python
dataset_folder = 'dataset/icare1.0/training'
fft_data = []
for patient_folder in os.listdir(dataset_folder):
    print(patient_folder)
    patient_path = os.path.join(dataset_folder, patient_folder)
    if not os.path.isdir(patient_path):
        continue

    full_signal = None
    for file_name in os.listdir(patient_path):
        if not file_name.endswith(".mat"):
            continue
        # print(file_name)
        file_path = os.path.join(patient_path, file_name)

        mat_data = loadmat(file_path)
        cur_signal = mat_data.get('val')  # numpy array of shape (18, )

        # write code to concatenate data along the second axis and save it to full_signal
        if full_signal is None:
            full_signal = cur_signal
        else:
            try:
                full_signal = np.hstack((full_signal, cur_signal))
            except Exception as e:
                print(e)
                continue
    patient_fft = perform_fft(full_signal)  # Size (18, 50) fixed
    fft_data.append(patient_fft)
```

```python
fft_data = np.array(fft_data)
print(fft_data.shape)
np.save('fft_data.npy', fft_data)
```

## Get patient features from .txt files

```python
dataset_folder = 'dataset/icare1.0/training'
patient_features = []
for patient_folder in os.listdir(dataset_folder):
    patient_path = os.path.join(dataset_folder, patient_folder)
    if not os.path.isdir(patient_path):
        continue

    values = []
    for file_name in os.listdir(patient_path):
        if not file_name.endswith(".txt"):
            continue
        print(file_name)
        file_path = os.path.join(patient_path, file_name)

        with open(file_path, 'r') as file:
            for line in file:
                value = line.split(':')[-1].strip()
                if value == 'nan':
                    value = float('nan')
                elif value.isdigit():
                    value = int(value)
                values.append(value)
    assert len(values) == 9
    patient_features.append(values)
```

```python
patient_features_df = pd.DataFrame(patient_features,
                                   columns=['Patient', 'Age', 'Sex', 'ROSC', 'OHCA', 'VFib', 'TTM', 'Outcome', 'CPC'])
patient_features_df
```

```python
patient_features_df.to_csv('patient_features.csv', index=False)
```

## Create full dataset

```python
patient_fft_loaded = np.load('fft_data.npy')
```

```python
patient_fft_df = pd.DataFrame(patient_fft_loaded.reshape(patient_fft_loaded.shape[0], -1))
patient_fft_df
```

```python
data_df = pd.concat([patient_features_df, patient_fft_df], axis=1)
data_df
```

```python
data_df.to_csv('preprocessed_data.csv', index=False)
```
